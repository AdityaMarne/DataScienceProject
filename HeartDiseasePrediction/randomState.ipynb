```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Create a 2D classification dataset
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, 
                           n_clusters_per_class=1, random_state=42)

# Try with two different random_states
states = [1, 42]

plt.figure(figsize=(10, 4))

for i, state in enumerate(states):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=state)
    
    plt.subplot(1, 2, i + 1)
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='coolwarm', label='Train', marker='o')
    plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='coolwarm', edgecolors='k', label='Test', marker='s')
    plt.title(f"Random State = {state}")
    plt.legend()

plt.tight_layout()
plt.show()
```
